```python
import pandas as pd

listing_df = pd.read_csv("listing_final.csv")

# find the PropertyType column case-insensitively
property_type_col = next((c for c in listing_df.columns if c.lower() == "propertytype"), None)
if property_type_col is None:
    raise KeyError("Column 'PropertyType' not found in listing_final.csv")

print("original shape:", listing_df.shape)

residential_df = listing_df[
    listing_df[property_type_col].astype(str).str.lower() == "residential"
]

print("residential shape:", residential_df.shape)
```

```python
sold_df = pd.read_csv("sold_final.csv")

# find the PropertyType column case-insensitively
sold_property_type_col = next((c for c in sold_df.columns if c.lower() == "propertytype"), None)
if sold_property_type_col is None:
    raise KeyError("Column 'PropertyType' not found in sold_final.csv")

print("original shape:", sold_df.shape)

sold_residential_df = sold_df[
    sold_df[sold_property_type_col].astype(str).str.lower() == "residential"
]

print("residential shape:", sold_residential_df.shape)
```
